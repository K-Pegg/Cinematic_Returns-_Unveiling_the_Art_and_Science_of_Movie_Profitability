```python
#merging basics_ratings & tmdb
basics_ratings_tmdb = pd.merge(basics_ratings, tmdb, on='original_title', how='inner')
basics_ratings_tmdb
```

```python
basics_ratings_tmdb.drop(['Unnamed: 0_x'], axis=1)
```

```python
budget_gross = pd.merge(budget,
                       gross,
                       left_on='movie',
                       right_on='title',
                       how='inner')
budget_gross
```

```python
df = pd.merge(basics_ratings_tmdb, 
                  budget_gross,
                  left_on='primary_title',
                  right_on='movie',
                  how='inner')
df
```

```python
df.isna().sum()
```

```python
df_profit_months=df.loc(axis=1)['primary_title', 'release_date_y', 'genres', 'net_profit']
df_profit_months
```

```python
df_profit_months['release_month']=pd.to_datetime(df_profit_months['release_date_y']).dt.strftime('%B')
df_profit_months
```

```python
df_profit_months=df_profit_months.drop('release_date_y', axis=1)
df_profit_months
```

```python
df_profit_months=df_profit_months.drop_duplicates('primary_title')
df_profit_months
```

```python
df_profit_months.loc[df_profit_months.duplicated()]
```

```python
df_profit_months=df_profit_months.assign(genres=df_profit_months.genres.str.split(",")).explode('genres')
df_profit_months
```

```python
profit_month_horror=df_profit_months.loc[df_profit_months.genres=='Horror']
profit_month_horror
```

```python
profit_month_horror.release_month.value_counts()
```

```python
profit_month_horror.groupby('release_month')['net_profit'].sum()
```

```python
profit_month_horror.groupby('release_month')['net_profit'].median()
```

```python
profit_month_horror.groupby('release_month')['net_profit'].mean()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 
               'August', 'September', 'October', 'November', 'December']
sns.barplot(x=profit_month_horror.release_month, y=profit_month_horror.net_profit, order=month_order, palette='viridis')
plt.title('Mean Profit per Month')
plt.xlabel('Release Month')
plt.xticks(rotation=45)
plt.ylabel('Mean Profit (10 Million)')
#plt.show()

plt.savefig('../Images/Mean_Profit_Horror.jpg', dpi=300)
```
