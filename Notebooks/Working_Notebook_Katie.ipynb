```python
#merging basics_ratings & tmdb
basics_ratings_tmdb = pd.merge(basics_ratings, tmdb, on='original_title', how='left')
basics_ratings_tmdb
```

```python
budget_gross = pd.merge(budget,
                       gross,
                       left_on='movie',
                       right_on='title',
                       how='left')
budget_gross
```

```python
df = pd.merge(basics_ratings_tmdb, 
                  budget_gross,
                  left_on='primary_title',
                  right_on='movie',
                  how='left')
df
```

```python
df.isna().sum()
```

```python
df=df.drop(['Unnamed: 0_x_x', 'Unnamed: 0_y_x', 'original_language'], axis=1)
df
```

```python
df_profit_genres=df.loc(axis=1)['primary_title', 'genres', 'net_profit']
df_profit_genres
```
